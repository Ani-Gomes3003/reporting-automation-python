```python
import pandas as pd

# Step 1: Read Excel file
df = pd.read_excel("data/raw_data.xlsx")

# Step 2: Create productivity column
df["Productivity"] = df["Tasks Completed"] / df["Hours Worked"]

# Step 3: Create error rate column
df["Error Rate"] = df["Errors"] / df["Tasks Completed"]

# Step 4: Identify low performers (Productivity < 5)
df["Low Performer"] = df["Productivity"] < 5

# Show result
df
```

```python
# Export cleaned and automated report
output_path = "data/processed_report.xlsx"

df.to_excel(output_path, index=False)

print("Report generated successfully:", output_path)
```

```python
# Calculate summary metrics
total_tasks = df["Tasks Completed"].sum()
avg_productivity = df["Productivity"].mean()
avg_error_rate = df["Error Rate"].mean()
low_performers_count = df["Low Performer"].sum()

# Create summary dataframe
summary = pd.DataFrame({
    "Metric": [
        "Total Tasks Completed",
        "Average Productivity",
        "Average Error Rate",
        "Low Performers Count"
    ],
    "Value": [
        total_tasks,
        avg_productivity,
        avg_error_rate,
        low_performers_count
    ]
})

summary
```

```python
# Export both detailed report and summary into same Excel file

with pd.ExcelWriter("data/final_report.xlsx") as writer:
    df.to_excel(writer, sheet_name="Detailed Report", index=False)
    summary.to_excel(writer, sheet_name="Summary", index=False)

print("Final report generated successfully")
```
